```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#unique items
uniqueItemNames = purchase_data["Item Name"].nunique()
#average price
averagePrice = purchase_data["Price"].mean()
#number of purchases
rowCount = purchase_data["Purchase ID"].count()
#Total rev
totalRevenue = purchase_data["Price"].sum()

data = {'Number of Unique Items' : [uniqueItemNames], 
     'Average Price': [averagePrice],
        'Number of Puchases': [rowCount],
        'Total Revenue': [totalRevenue]
       } 
  
# create the dataframe 
dataframe2 = pd.DataFrame(data, columns = ['Number of Unique Items', 'Average Price','Number of Puchases','Total Revenue'])
pd.options.display.float_format = '${0:,.2f}'.format

#print(dataframe2)
dataframe2.head()


```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
newDataFrame = purchase_data[['SN','Gender']]

#counting up the genders
SN_type = newDataFrame[['SN','Gender']].drop_duplicates()
maleCount = SN_type['Gender'].value_counts()['Male']
femaleCount = SN_type['Gender'].value_counts()['Female']
otherCount = SN_type['Gender'].value_counts()['Other / Non-Disclosed']

#getting number of unique values
index = SN_type.index
number_of_rows = len(index)

#gender percentages
malePercent = maleCount/number_of_rows
femalePercent = femaleCount/number_of_rows
otherPercent = otherCount/number_of_rows

#dictionary of information
data = pd.DataFrame({'':['Male','Female','Other / Non-Disclosed'],
                     'Total Count': [maleCount,femaleCount,otherCount],
                     'Percentage of Players': [malePercent, femalePercent, otherPercent],
                    })

#print and format data
dataframe3 = pd.DataFrame(data,columns = ['','Total Count','Percentage of Players'])
pd.options.display.float_format = "{0:.2%}".format

dataframe3.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
newDataFrame2 = purchase_data[['Gender','Price','SN']]

#purchase count
maleCount2 = newDataFrame2['Gender'].value_counts()['Male']
femaleCount2 = newDataFrame2['Gender'].value_counts()['Female']
otherCount2 = newDataFrame2['Gender'].value_counts()['Other / Non-Disclosed']

#calculating the average purchase price per gender
femaledf = newDataFrame2.query("Gender=='Female'")['Price'].mean()
maledf = newDataFrame2.query("Gender=='Male'")['Price'].mean()
otherdf = newDataFrame2.query("Gender=='Other / Non-Disclosed'")['Price'].mean()

#calculate the total purchase value 
femalesum = newDataFrame2.query("Gender=='Female'")['Price'].sum()
malesum = newDataFrame2.query("Gender=='Male'")['Price'].sum()
othersum = newDataFrame2.query("Gender=='Other / Non-Disclosed'")['Price'].sum()

#calculate the average total purchase per person
avPurchase = newDataFrame2.groupby(['Gender','SN']).sum()
femaleAvTotal = avPurchase.query("Gender=='Female'")['Price'].mean()
maleAvTotal = avPurchase.query("Gender=='Male'")['Price'].mean()
otherAvTotal = avPurchase.query("Gender=='Other / Non-Disclosed'")['Price'].mean()


#store data in dictionary
data2 = pd.DataFrame({'Gender':['Female','Male','Other / Non-Disclosed'],
                      'Purchase Count': [femaleCount2,maleCount2,otherCount2],
                      'Average Purchase Price': [femaledf, maledf, otherdf],
                      'Total Purchase Value':[femalesum,malesum,othersum],
                      'Average Total Purchase per Person': [femaleAvTotal, maleAvTotal, otherAvTotal],
                    })

#print and format data
dataframe4 = pd.DataFrame(data2,columns = ['Gender','Purchase Count','Average Purchase Price','Total Purchase Value','Average Total Purchase per Person'])
pd.options.display.float_format = '${0:,.2f}'.format

dataframe4.head()



```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0,9,14,19,24,29,34,39,100]
age_labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
age_data = purchase_data[['SN','Age']]

#get unique ids count
num_rows = purchase_data['SN'].nunique()

#bin data 
age_data['Age Range']= pd.cut(age_data['Age'], age_bins, labels = age_labels)
age_data = age_data.drop_duplicates('SN')

#count and calculate
age_data = age_data[['Age Range','Age']]
age_data = age_data.groupby('Age Range').count()
age_data['Percentage of Players'] = (age_data / num_rows)*100

#format
pd.options.display.float_format = '{0:,.2f}%'.format

#rename
age_data.rename(columns={'Age':'Total Count'},inplace=True)

age_data


```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purch_ana_data = purchase_data[['SN','Age','Price']]
purch_ana_data.rename(columns={'Age':'Purchase Count'},inplace=True)
purch_ana_data['Age Range']= pd.cut(purch_ana_data['Purchase Count'], age_bins, labels = age_labels)
purch_ana_data

#count groups
age_count = purch_ana_data[['Age Range','Purchase Count']]
age_count = age_count.groupby('Age Range').count()

#average purchase count
av_purch_price = purch_ana_data[['Age Range','Price']]
av_purch_price = av_purch_price.groupby('Age Range').mean()
av_purch_price

#Total purchase value
tot_purch_val = purch_ana_data[['Age Range','Price']]
tot_purch_val = tot_purch_val.groupby('Age Range').sum()
tot_purch_val

#average total purchase per person 
sum_tot_purch = purch_ana_data[['SN','Price']]
sum_tot_purch = sum_tot_purch.groupby(['SN']).sum()
sum_tot_purch.rename(columns={'Price':'Sum'},inplace=True)
sum_tot_purch

# sum_tot_purch
avg_tot_purch = purch_ana_data[['Age Range','SN','Price']]
avg_tot_purch = avg_tot_purch.merge(sum_tot_purch,left_on='SN',right_on='SN')
avg_tot_purch = avg_tot_purch.drop_duplicates(subset='SN')
avg_tot_purch = avg_tot_purch[['Age Range','Sum']]
avg_tot_purch = avg_tot_purch.groupby('Age Range').mean()

# Merge dfs
age_count = age_count.merge(av_purch_price,left_on='Age Range',right_on='Age Range')
age_count.rename(columns={'Price':'Average Purchase Price'},inplace=True)
age_count = age_count.merge(tot_purch_val,left_on='Age Range',right_on='Age Range')
age_count.rename(columns={'Price':'Total Purchase Value'},inplace=True)
age_count = age_count.merge(avg_tot_purch,left_on='Age Range',right_on='Age Range')
age_count.rename(columns={'Sum':'Avg total Purchase per Person'},inplace=True)

age_count
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
summary_frame = purchase_data[['SN','Price']]
count_frame = summary_frame

#count
count_frame = count_frame.groupby('SN').count()

#average price
av_price = summary_frame
av_price = av_price.groupby('SN').mean()

#total purchase 
tot_purch = summary_frame
tot_purch = tot_purch.groupby('SN').sum()

#merge dfs
count_frame = count_frame.merge(av_price,left_on='SN',right_on='SN')
count_frame = count_frame.merge(tot_purch,left_on='SN',right_on='SN')

#rename columns
count_frame.rename(columns={'Price_y':'Purchase Count'},inplace=True)
count_frame.rename(columns={'Price':'Total Purchase Value'},inplace=True)
count_frame.rename(columns={'Price_x':'Purchase Count'},inplace=True)

#sort
count_frame = count_frame.sort_values(by=['Total Purchase Value'],ascending=False)

#print
count_frame
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_pop = purchase_data[['Item ID','Item Name','Price']]

#count
most_pop_count = most_pop[['Item ID','Item Name']]
most_pop_count = most_pop_count.groupby('Item ID').count()

#total purchase value
most_pop_purch_val = most_pop[['Item ID','Price']]
most_pop_purch_val = most_pop_purch_val.groupby('Item ID').sum()

#merge
most_pop = most_pop.merge(most_pop_count,left_on='Item ID',right_on='Item ID')
most_pop = most_pop.merge(most_pop_purch_val,left_on='Item ID',right_on='Item ID')

#rename columns
most_pop.rename(columns={'Item Name_y':'Purchase Count'},inplace=True)
most_pop.rename(columns={'Price_y':'Total Purchase Value'},inplace=True)
most_pop.rename(columns={'Price_x':'Item Price'},inplace=True)
most_pop.rename(columns={'Item Name_x':'Item Name'},inplace=True)

#sort
most_pop = most_pop.sort_values(by=['Item Price'],ascending=True)
most_pop = most_pop.drop_duplicates(subset=['Item ID'],keep='first')
most_pop = most_pop.sort_values(by=['Purchase Count'],ascending=False)
most_pop = most_pop[['Item ID','Item Name','Purchase Count','Item Price','Total Purchase Value']]

most_pop
```
